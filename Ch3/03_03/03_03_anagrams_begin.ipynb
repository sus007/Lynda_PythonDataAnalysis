```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))

def signature(word):
    return ''.join(sorted(word))

# signature('Land')

def anagram(myword):
    return [word for word in wordclean if signature(myword)==signature(word)]

anagram('life')
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
from collections import defaultdict
ourList = defaultdict(list)
for word in wordclean:
    ourList[signature(word)].append(word)

def newAnagram(targetword):
    return ourList[signature(targetword)]

newAnagram('life')
# Ignoring the specific words which are the anagrams of themselves, i.e the onecharacter words.
%timeit anagram_all = {word: newAnagram(word) for word in wordclean if len(newAnagram(word)) > 1}
```

```python
# Challange 1
# Step 1- Importing the dictionary
wordList = sorted(list(set([word.strip().lower() for word in open('words', 'r')])))

# Step 2- Importing defaultdict constructor from collections module.
from collections import defaultdict
lengthSet = defaultdict(list) # creates a default dict. which has list as default value containers.

# Step 3- Arranging words of specific length into specific classes.
for aWord in wordList:
    lengthSet[len(aWord)].append(aWord) #the key for lengthSet dict. has been set as the respective number('digit').

# # Step 4- Finding all anagrams for each class.
anagramCount = {} # initializing anagramCount dict outside the loop to avoid reinitialization in every iteration.
for className, valueList in lengthSet.items(): 
    anagramList = defaultdict(list)
    for word in valueList:
        if len(word)>1:
            anagramList[className].append(newAnagram(word))
        else:
            continue;
    anagramCount.update({className: int(len(anagramList[className])/2.0)})#[className]=len(anagramList[className])

# Step 5- print out the final dict containing the numbers of anagrams of respective classes.    
anagramCount
```
